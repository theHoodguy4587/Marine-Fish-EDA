```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
fish=pd.read_csv('Marine_Fish_Data.csv')
fish.head()
```

```python
pd.isnull(fish).sum()
```

```python
# 1.General Data Understanding
```

```python
#How many unique fish species are in the dataset?
```

```python
fish['Species_Name'].nunique()
```

```python
#What is the distribution of fish populations across regions?
```

```python
sns.barplot(x='Region',y='Fish_Population',data=fish)
```

```python
#Which fishing methods are most commonly used?
```

```python
sns.countplot(data=fish,x='Fishing_Method')
```

```python
# 2. Population and Size Trends
```

```python
#What is the distribution of fish population sizes? Are there any extreme values?
```

```python
sns.boxplot(data=fish,x='Species_Name',y='Fish_Population')
```

```python
#What is the average size of fish in each species?
```

```python
sns.barplot(data=fish,x='Species_Name',y='Average_Size(cm)')
```

```python
#Are there specific regions with significantly higher or lower fish populations?
```

```python
sns.barplot(data=fish,x='Region',y='Fish_Population')
```

```python
# 3. Environmental Impact
```

```python
#How does water pollution level vary across regions?
```

```python
sns.countplot(data=fish,x='Region',hue='Water_Pollution_Level')
```

```python
# What is the relationship between water temperature and fish population size?
```

```python
sns.lmplot(data=fish,x='Water_Temperature(C)',y='Fish_Population')
plt.tight_layout()
```

```python
#Is there a noticeable trend in fish population size across different water pollution levels?
```

```python
sns.boxplot(data=fish,x='Water_Pollution_Level',y='Fish_Population')
```

```python
# 4. Overfishing Risks
```

```python
# Which species are most at risk of overfishing?
```

```python
sns.countplot(data=fish,x='Species_Name',hue='Overfishing_Risk')
```

```python
# How does overfishing risk vary with fishing methods?
```

```python
sns.countplot(x='Fishing_Method',data=fish,hue='Overfishing_Risk')
```

```python
# Is there a link between pollution levels and overfishing risk?
```

```python
sns.countplot(x='Water_Pollution_Level',data=fish,hue='Overfishing_Risk')
```

```python
# 5. Seasonal and Regional Patterns
```

```python
# How does the average fish size vary by breeding season?
```

```python
sns.boxplot(x='Breeding_Season',y='Average_Size(cm)',data=fish,hue='Breeding_Season')
```

```python
sns.lineplot(x='Breeding_Season',y='Fish_Population',data=fish)
```

```python
# Which regions show the highest or lowest risk of overfishing?
```

```python
sns.heatmap(fish.pivot_table(values='Fish_Population',index='Region',columns='Overfishing_Risk'))
```

```python
# 6. Correlations and Interactions
```

```python
# Are larger fish more at risk of overfishing?
```

```python
sns.boxplot(data=fish,y='Average_Size(cm)',x='Overfishing_Risk')
```

```python
# Is there a correlation between water temperature and the average size of fish?
```

```python
sns.lmplot(x='Water_Temperature(C)',y='Average_Size(cm)',data=fish)
```

```python
# Heatmap for correlations
```

```python
correlated=fish[['Fish_Population','Average_Size(cm)','Water_Temperature(C)']].corr()
```

```python
sns.heatmap(correlated,cmap='coolwarm',annot=True)
```
